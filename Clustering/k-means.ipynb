```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
data=pd.read_csv('/content/drive/MyDrive/Machine Learning/Clustering/Salary (1).csv')
```

```python
data
```

```python
data.head()
```

```python
plt.scatter(data.YearsExperience,data['Salary'])
plt.xlabel('Years experience')
plt.ylabel('salary')
```

```python
km=KMeans(n_clusters=3)
y_predicted=km.fit_predict(data[['YearsExperience','Salary']])
y_predicted
```

```python
data['cluster']=y_predicted
data.head()
```

```python
data1 = data[data.cluster==0]
data2 = data[data.cluster==1]
data3 = data[data.cluster==2]
plt.scatter(data1.YearsExperience,data1['Salary'],color='green')
plt.scatter(data2.YearsExperience,data2['Salary'],color='red')
plt.scatter(data3.YearsExperience,data3['Salary'],color='black')
# plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Years Experience')
plt.ylabel('Salary')
plt.legend()
```

```python
# Preprocessing using min max scaler
scaler = MinMaxScaler()

scaler.fit(data[['Salary']])
data['Salary'] = scaler.transform(data[['Salary']])

scaler.fit(data[['YearsExperience']])
data['YearsExperience'] = scaler.transform(data[['YearsExperience']])
plt.scatter(data.YearsExperience,data['Salary'])
```
