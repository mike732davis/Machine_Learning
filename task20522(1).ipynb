```python
for i in ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides','pH','sulphates']:
  data2[i]=data2[i].fillna(data2[i].median())
```

```python
data2.isna().sum()
```

```python
plt.boxplot(data2['fixed acidity'])
plt.title('fixed acid')
```

```python
q1 = np.percentile(data2['fixed acidity'],25, interpolation='midpoint')
q2 = np.percentile(data2['fixed acidity'],50, interpolation='midpoint')
q3 = np.percentile(data2['fixed acidity'],75, interpolation='midpoint')
print(q1,q2,q3)
```

```python
data2['fixed acidity'].median()
```

```python
iqr=q3-q1
low_limit=q1-1.5*iqr
upper_limit=q3+1.5*iqr
print(upper_limit,low_limit)
```

```python
outlier=[]
for x in data2['fixed acidity']:
  if ((x>upper_limit) or (x<low_limit)):
    outlier.append(x)
outlier
```

```python
ind1=data2['fixed acidity']>upper_limit
data2.loc[ind1].index
```

```python
a=data2.loc[ind1].index
data2.drop(a,inplace=True)
```

```python
plt.boxplot(data2['fixed acidity'])
plt.title('fixed acid')
```
