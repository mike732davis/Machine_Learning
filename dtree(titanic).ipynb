```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
data=pd.read_csv('/content/drive/MyDrive/Machine Learning/titanic.csv')
```

```python
data.head()
```

```python
data_n=data.drop(['PassengerId','Name','Ticket','SibSp','Parch','Fare','Cabin','Embarked'],axis='columns')
```

```python
data_n.head()
```

```python
input=data_n.drop('Survived',axis='columns')
target=data_n.Survived
```

```python
input.Sex=input.Sex.map({'male':1,'female':2})
input.Age[:10]
```

```python
input.Age=input.Age.fillna(input.Age.mean())
input.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(input, target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(X_train,Y_train)
```

```python
model.score(X_test, Y_test)
```

```python
model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(Y_test, model.predict(X_test)))
print(confusion_matrix(Y_test, model.predict(X_test))) 
```

```python
from sklearn.ensemble import RandomForestClassifier
x1=RandomForestClassifier(n_estimators=100)
x1.fit(X_train, Y_train)
```

```python
x1.predict(X_test)
```

```python
print(classification_report(Y_test, x1.predict(X_test)))
print(confusion_matrix(Y_test, x1.predict(X_test)))
```
