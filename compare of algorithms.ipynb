```python
import pandas as pd
import numpy as np
import seaborn as sns
data=pd.read_csv('/content/drive/MyDrive/Machine Learning/iris.csv')
```

```python
data
```

```python
x=data[['Sepal_Length','Sepal_Width','Petal_Length','Petal_Width']]
y=data.Class
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test= train_test_split(x,y, test_size=0.2, random_state=1)
```

```python
len(X_train)
```

```python
len(Y_train)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
knn.fit(X_train,Y_train)
```

```python
knn.score(X_test,Y_test)
```

```python
knn.predict([[4.8,3.5,1.2,0.2]])
```

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(X_train,Y_train)
```

```python
dtree.predict([[4.8,3.5,1.2,0.2]])
```

```python
from sklearn.ensemble import RandomForestClassifier
ranfor=RandomForestClassifier()
ranfor.fit(X_train,Y_train)
```

```python
ranfor.predict([[4.8,3.5,1.2,0.2]])
```

```python
knn_pred=knn.predict(X_test)
dtree_pred=dtree.predict(X_test)
ranfor_pred=ranfor.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(Y_test,knn_pred))
print(confusion_matrix(Y_test,knn_pred))
```

```python
print(classification_report(Y_test,dtree_pred))
print(confusion_matrix(Y_test,dtree_pred))
```

```python
print(classification_report(Y_test,ranfor_pred))
print(confusion_matrix(Y_test,ranfor_pred))
```
