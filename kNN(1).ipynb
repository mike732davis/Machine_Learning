```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('/content/drive/MyDrive/Machine Learning/insurance_data.csv')
```

```python
data.head()
```

```python
plt.xlabel('affordability')
plt.ylabel('bought insurance')
plt.scatter(data['age'],data['affordibility'],color='red',marker='*')
plt.scatter(data['age'],data['bought_insurance'],color='blue',marker='+')
```

```python
from sklearn.model_selection import train_test_split
X = data.age
y = data.bought_insurance
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
len(X_train)
```

```python
len(X_test)
```

```python
len(y_train)
```

```python
len(y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)
```

```python
X_train=X_train.values.reshape(-1,1)
X_test=X_test.values.reshape(-1,1)
```

```python
knn.fit(X_train, y_train)
```

```python
knn.score(X_test,y_test)
```

```python
knn.predict([[25]])
```

```python
knn.predict([[50]])
```

```python
from sklearn.metrics import classification_report as rep 
y_predict=knn.predict(X_test)
print(rep(y_test,y_predict))
```
